### **COVID 19**

 - 2020년 1월 30일 ~ 2023년 5월 5일
 - 2020년 2월로 설정


#### **코로나 팬데믹 기간 여부 T/F**

```python
df_covid = df.copy()[['Year','Month']]
```

```python
condition1 = (df_covid['Year'] == 2020) & (df_covid['Month'] >= 2)
condition2 = (df_covid['Year'] > 2020) & (df_covid['Year'] < 2023)
condition3 = (df_covid['Year'] == 2023) & (df_covid['Month'] <= 5)

mask = condition1 | condition2 | condition3

df_covid['Pandemic'] = False
df_covid.loc[mask, 'Pandemic'] = True
```

```python
df_covid.loc[df_covid['Year']==2023]
```

```python
df_covid.to_csv('/content/drive/MyDrive/Data/Covid.csv', index=False)
```

#### **코로나 감염자수**

```python
n_covid = pd.read_excel('/content/drive/MyDrive/Causal_Inference/rawdata/covid.xlsx')
```

```python
df_n_covid = df.copy()[['Year','Month']]
```

```python
n_covid=n_covid.iloc[5:,:2]
n_covid.columns=['Date','n_covid']

n_covid.reset_index(drop=True, inplace=True)

n_covid['Date'] = pd.to_datetime(n_covid['Date'])
n_covid['Year'] = n_covid['Date'].dt.year
n_covid['Month'] = n_covid['Date'].dt.month
```

```python
n_covid.head()
```

```python
monthly_covid_cases = n_covid.groupby(['Year','Month'])['n_covid'].sum()
df_n_covid = df_n_covid.merge(monthly_covid_cases, on=['Year','Month'], how='left')
df_n_covid.fillna(0, inplace=True)
```

```python
df_n_covid.to_csv('/content/drive/MyDrive/Causal_Inference/Data/Covid19.csv', index=False)
```

### **GDP**


```python
# 프레임 준비
df_GDP = df.copy()[['Year','Month']]
```

```python
# GDP 데이터 불러오기
raw_GDP = pd.read_excel('/content/drive/MyDrive/rawdata/gdp.xlsx')

# 정리
temp_GDP=raw_GDP.iloc[1:4, 1:]
temp_GDP.columns = temp_GDP.iloc[0]
temp_GDP = temp_GDP[1:].reset_index(drop=True)
```

```python
df_gdp_value = pd.DataFrame(temp_GDP.iloc[0]).reset_index()
df_gdp_value.columns = ['Quarter', 'GDP']
df_gdp_value.head()
```

```python
# 각 행을 3번씩 반복해 월별 데이터로 생성
df_gdp_monthly = df_gdp_value.loc[df_gdp_value.index.repeat(3)].reset_index(drop=True)
```

```python
df_GDP['GDP']=df_gdp_monthly['GDP']
```

```python
df_GDP.head()
```

```python
df_GDP.to_csv('/content/drive/MyDrive/Data/GDP.csv', index=False)
```
